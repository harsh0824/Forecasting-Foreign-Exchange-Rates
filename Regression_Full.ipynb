```python
#importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.api as sm
import seaborn as sn
import sklearn
import math
```

```python
#Reading the data CSV
data = pd.read_csv("BDT_Data.csv")
```

```python
#the data looks like
data.head()
```

# Lets see if there is any relation between Bangladeshi Inflation rate and the exchange rate

```python
#Lets see what is the equation of the line
X = data['BIR'].values.reshape(-1,1)
y = data['ER'].values.reshape(-1,1)
reg = LinearRegression()
reg.fit(X, y)
print("The linear model is: Y = {:.5} + {:.5}X".format(reg.intercept_[0], reg.coef_[0][0]))

#Let us look at the effect of Bangladesh Inflation rate on the Exchange rate
predictions = reg.predict(X)
plt.figure(figsize=(16, 8))
plt.scatter(
    data['BIR'],
    data['ER'],
    c='black'
)
plt.plot(
    data['BIR'],
    predictions,
    c='blue',
    linewidth=2
)
plt.xlabel("Bangladeshi inflation rate(%)")
plt.ylabel("Exchange rate")
plt.show()

# Getting some data like p and f statistics

X = data['BIR']
y = data['ER']
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```

# # From the comparison of the R squared and p values from the above analysis, it seems reasonable to accept BIvW, GDP_B and GBvW are good indicators/determiners for the exchange rate.

### From the correlation matrix, we can conclude that since BIR and BIvU have been removed, we wont have highly correlated determiners left. But there is one issue, GDP_B and GBvW are highly correlated. So we will test two three models, BIvW will be there in all but the GDP_B and GBvW will be present alternatively. Then we will test to find out the best one.

## Model 1: Including only BIvW and GDP_B

```python
Xs = data.drop(['Year','ER','BIR','UIR','WIR','BIvU','ReIRB','ReIRU','RIRBvU','GDP_W','GBvW','BOP'], axis=1)
y = data['ER'].values.reshape(-1,1)
reg = LinearRegression()
reg.fit(Xs, y)
print("Model: Y = {:.5} + {:.5}*BIvW + {:.5}*GDP_B ".format(reg.intercept_[0], reg.coef_[0][0], reg.coef_[0][1]))


X = np.column_stack((data['BIvW'],data['GDP_B'],))
y = data['ER']
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```

## Model 2: BIvW and GBvW

```python
Xs = data.drop(['Year','ER','BIR','UIR','WIR','BIvU','ReIRB','ReIRU','RIRBvU','GDP_B','GDP_W','BOP'], axis=1)
y = data['ER'].values.reshape(-1,1)
reg = LinearRegression()
reg.fit(Xs, y)
print("Model: Y = {:.5} + {:.5}*BIvW + {:.5}*GBvW ".format(reg.intercept_[0], reg.coef_[0][0], reg.coef_[0][1]))


X = np.column_stack((data['BIvW'],data['GBvW'],))
y = data['ER']
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```

## Model 3: BIvW, GDP_B and GBvW

```python
Xs = data.drop(['Year','ER','BIR','UIR','WIR','BIvU','ReIRB','ReIRU','RIRBvU','GDP_W','BOP'], axis=1)
y = data['ER'].values.reshape(-1,1)
reg = LinearRegression()
reg.fit(Xs, y)
print("Model: Y = {:.5} + {:.5}*BIvW + {:.5}*GDP_B + {:.5}*GBvW ".format(reg.intercept_[0], reg.coef_[0][0], reg.coef_[0][1],reg.coef_[0][2]))


X = np.column_stack((data['BIvW'],data['GDP_B'],data['GBvW']))
y = data['ER']
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```

## All of the three models have performed quite well. Let us now forecast using these models and some tests.
